```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
path = 'D:/AAA/chromedriverrr.exe'
```

```python
title = []
loc = []
desc = []
price = []
rooms = []
owner = []
```

```python
for i in range(1,20,1):
    main_link = "https://www.lamudi.co.id/jakarta/apartment/buy/?q=apartemen+jakarta&page=" + str(i)
    
    #customize chrome display
    chrome_option = Options()
    chrome_option.add_argument('--no-sandbox')
    chrome_option.add_argument('disable-notification')
    chrome_option.add_argument('--disable-infobars')
    
    driver = webdriver.Chrome(executable_path = path, options=chrome_option)
    driver.get(main_link)
    
    html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    soup = BeautifulSoup(html,"html.parser")
    
    #Looping Dataset
    
    #Title
    for i in soup.find_all('h2', class_="ListingCell-KeyInfo-title"):
        title.append(i.text)
       
    #Location
    for i in soup.find_all('span', class_="ListingCell-KeyInfo-address-text"):
        loc.append(i.text)
    
    #Description
    for i in soup.find_all('div', class_="ListingCell-shortDescription"):
        desc.append(i.text)
    
    #Price
    for i in soup.find_all('span', class_="PriceSection-FirstPrice"):
        price.append(i.text)
    
    #Rooms
    for i in soup.find_all('div', class_="ListingCell-keyInfo-details"):
        rooms.append(i.text)
    
    #Owner
    for i in soup.find_all('div', class_="ListingDetail-agent-name"):
        owner.append(i.text)
```

```python
print(len(title))
print(len(loc))
print(len(desc))
print(len(price))
print(len(rooms))
print(len(owner))
```

```python
#Save Data

listcols = ['title', 'location', 'description', 'price', 'rooms', 'owner']
dict_data = dict(zip(listcols,(title,loc,desc,price,rooms,owner)))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dict_data)
df
```
